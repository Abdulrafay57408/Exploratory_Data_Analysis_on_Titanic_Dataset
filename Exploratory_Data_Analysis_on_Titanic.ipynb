```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


csv_file = pd.read_csv(r"/train.csv")
csv_file.head()
```

```python
null_values =  csv_file.isnull()
null_values
```

```python

sum_of_null_values = csv_file.isnull().sum()
sum_of_null_values
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(null_values)
```

```python

percent_of_null_values = sum_of_null_values / csv_file.shape[0] * 100
percent_of_null_values
```

```python
csv_file.head()
```

```python
age_mean = csv_file["Age"].mean()
age_mean
```

```python

csv_file['Age'].fillna(value= age_mean , inplace= True )
csv_file.isnull().sum()
```

```python
# Put the mean of cabin at the null places of cabin column

csv_file.drop(columns=['Cabin'], inplace=True)
csv_file.isnull().sum()
```

```python
csv_file = csv_file.dropna()
csv_file.isnull().sum()
```

```python
# plot heatmap

plt.figure(figsize=(20,20))
sns.heatmap(csv_file.isnull())
```

```python
csv_file.head()
```

```python
sex = pd.get_dummies(csv_file['Sex'] , drop_first=True).astype(int)
sex
```

```python
embark = pd.get_dummies( csv_file['Embarked'], drop_first = True).astype(int)
embark
```

```python
sns.countplot(x='Survived', hue='Sex', data=csv_file)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=csv_file)
plt.title('Survival Rate by Class')
plt.show()
```

```python
sns.barplot(x = 'Survived' , hue='Pclass' , data= csv_file)
plt.title("Survival Rate by Class")
plt.show()
```

```python
sns.kdeplot(csv_file[csv_file['Survived'] == 1]['Age'], label='Survived', fill=True)
sns.kdeplot(csv_file[csv_file['Survived'] == 0]['Age'], label='Not Survived', fill=True)
plt.title('Age Distribution by Survival')
plt.legend()
plt.show()
```

```python
csv_file['FamilySize'] = csv_file['SibSp'] + csv_file['Parch'] + 1
csv_file.head()
```

```python
sns.barplot(x='FamilySize', y='Survived', data=csv_file)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
csv_file.drop(['Sex', 'Embarked' , 'Ticket'], axis=1 , inplace=True)
csv_file.head()
```

```python
csv_file = pd.concat([csv_file , sex , embark], axis=1)
csv_file.head()
```

```python
sum_of_null_values = csv_file.isnull().sum()
sum_of_null_values
```

```python
sns.countplot(x='male', data=csv_file)
plt.title('Gender Classification')
plt.show()
```

```python
sns.barplot(x='Survived', hue='male', data=csv_file)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.countplot(x='Survived', hue='male', data=csv_file)
plt.title('Survival Rate By Gender')
plt.show()
```
